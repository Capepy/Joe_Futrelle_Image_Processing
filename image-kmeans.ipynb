```python
%matplotlib inline
from matplotlib import pyplot as plt
from utils import show_image, show_mask, show_channels
from skimage.io import imread

# using K-means to quantize an image to a reduced set of colors

rgb = imread('poster.png')
show_image(rgb)
# note the subtle canvas texture throughout this image; hopefully K-means will ignore it
```

```python
import numpy as np
from scipy.cluster.vq import kmeans2

# get the height and width of the image, c is the number of color channels (3)
h, w, c = rgb.shape

# create an N x 3 feature vector where N is the height * width of the image
pixels = rgb.reshape((h*w,3))
# now compute K means using K-means. Select K=6 because that is how
# many colors the image contains, when you ignore the canvas texture
K = 6
means, labels = kmeans2(pixels,K)

# "means" are the K means that k-means found.
# K-means starts with random means, so sometimes it generates
# unfortunate results (solving that is an application-specific problem)
means
```

```python
# labels are, for each of the N pixels, which mean is the closest.
# this visualization just assigns each mean a grayscale value
# based on the order the means came back from K-means, which is random,
# so the intensity values shown here do not correspond to the original image
show_image(labels.reshape(h,w) / float(K))
```

```python
# to visualize the color of the means, first convert back to unsigned integer pixel values
mi = means.astype(np.uint8)
# now construct a 1xK pixel multi-channel image of the means
plt.imshow(mi.reshape((1,K,3)),interpolation='nearest')
```

```python
# now select, for the mean associated with each pixel, its color
# do it for each channel
channels = [np.take(means[:,c],labels).reshape(h,w) for c in range(3)]
# merge the channels into a h x w x 3 color image
out = np.dstack(channels)
show_image(out.astype(np.uint8))
# note that the subtle canvas texture has been ignored,
# and the colors have, for the most part, been correctly
# identified
```

```python
# for illustrative purposes, generate a mask from the location
# of one of the means 
m=3
li = labels.reshape((h,w))
cmask = li == m
show_mask(cmask)
```
