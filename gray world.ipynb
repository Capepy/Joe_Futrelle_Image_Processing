```python
%matplotlib inline
from matplotlib import pyplot as plt
from utils import show_image, show_mask, show_channels
from skimage.io import imread

"""
Gray world: automatic color balancing based on average image color
"""

crab = imread('crab.png')
show_image(crab)
```

```python
import numpy as np
from skimage import img_as_float

# compute mean of each color channel
a=img_as_float(crab)
np.mean(a.reshape(-1,3),axis=0)
```

```python
# now make a 1x1 pixel image that color
means = np.mean(a.reshape(-1,3),axis=0)
plt.imshow(means.reshape(1,1,3))
```

```python
# the gray point is the mean of the means of all the color channels
gray = np.mean(means)

# gray world algorithm is image * (gray / means)
wb = (a * (gray / means)).clip(0,1)
# result is color-corrected multichannel image
show_image(wb)
```
