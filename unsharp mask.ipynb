```python
from skimage.io import imread

# unsharp mask: increase local contrast of image

rgb = imread('benthic.png')
show_image(rgb)
```

```python
from skimage.color import rgb2gray
from skimage import img_as_ubyte

# convert image to grayscale.
# note that this does not just take the mean of the channels;
# red, green, and blue are weighted by empirically-derived factors
y = rgb2gray(rgb)

show_image(y)
```

```python
from skimage.filter import gaussian_filter

# the first step in unsharp masking is to filter the image to remove
# high spatial frequency features, using a Gaussian filter

# arbitrarily select a Gaussian filter size of 10 pixels; this is one
# of the parameters of the unsharp mask algorithm
show_image(gaussian_filter(y,10))
```

```python
# here's the whole algorithm. in effect the blurred image is subtracted
# from the original image to leave only the higher-frequency features.
# some scaling needs to be applied, controlled by a second parameter
def unsharp_mask(image,size,a=2):
    a = float(a)
    um = gaussian_filter(image,size)
    return (a / (a - 1)) * (image - um / a)

show_image(unsharp_mask(y,10,2).clip(0,1))
```

```python
# let's try a different value for a. values less than 1 should not be used;
# values close to 1 give highly exaggerated results. very large values
# result in almost no effect on the original image

# another thing to note about unsharp mask is that it amplifies noise, which
# tends to have a high spatial frequency

show_image(unsharp_mask(y,10,1.3).clip(0,1))
```

```python
# in color, unsharp masking is applied to each channel separately
import numpy as np
from skimage import img_as_float

rgb_float = img_as_float(rgb)
rgb_um = np.dstack([unsharp_mask(rgb_float[:,:,c],10,2).clip(0,1) for c in range(3)])
show_image(rgb_um)
```
