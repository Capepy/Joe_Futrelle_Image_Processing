```python
%matplotlib inline
from utils import show_image, show_mask, show_channels
```

```python
from skimage.io import imread

# locate a colorful object on a random background

rgb = imread('starfish.png')
show_image(rgb)
```

```python
%matplotlib inline

# show the rgb channels. note how strong the starfish is on the red channel
show_channels(rgb)
```

```python
from skimage.color import rgb2hsv

# now convert the image to hue, saturation, and value channels
# this shows that the starfish is bright and saturated
hsv = rgb2hsv(rgb)
show_channels(hsv)
```

```python
# split the channels into invidual grayscale images
R, G, B = [rgb[:,:,c] for c in range(3)]
H, S, V = [hsv[:,:,c] for c in range(3)]
```

```python
from skimage.filter import threshold_otsu

# now use Otsu's thresholding method

def thresh(y):
    return y > threshold_otsu(y)

# look for objects that are especially red, saturated, and bright
# use logical and to discard pixels that don't meet all three criteria
B = thresh(R) & thresh(S) & thresh(V)
show_mask(B)
# the starfish is the largest object in this mask
```

```python
from scipy.ndimage.measurements import label
from skimage.color import label2rgb

# now label all the individual connected regions in the image
L, _ = label(B)
# display each region as a different color
show_image(label2rgb(L,bg_label=0,bg_color=[0,0,0]))
```

```python
import numpy as np
from skimage.measure import regionprops

# find the label of the region with the greatest area
rp = regionprops(L)
largest = np.argmax([r.area for r in rp]) + 1
largest
```

```python
# now mask everything but that region
seg = L==largest
show_mask(seg)
```

```python
from scipy.ndimage.morphology import binary_fill_holes

# use morphological operation to fill in all enclosed black spaces in the mask
filled = binary_fill_holes(seg)
show_mask(filled)
```

```python
from skimage.segmentation import mark_boundaries

# now we have an estimate of the location of the starfish;
# visualize its boundary overlaid on the image
show_image(mark_boundaries(rgb,filled))
```
