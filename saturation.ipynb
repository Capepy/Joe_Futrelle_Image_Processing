```python
%matplotlib inline
from utils import show_image, show_mask, show_channels
```

```python
from skimage.io import imread

# using color models to increase the saturation of a color image

rgb = imread('astronaut.png')
show_image(rgb)
```

```python
%matplotlib inline

# here are the red, green, and blue channels of the image.
# note the differences

show_channels(rgb)
```

```python
from skimage.color import rgb2hsv

# now convert from RGB colorspace into hue, saturation, and value (HSV) space

hsv = rgb2hsv(rgb)

# show the hue, saturation, and value channels
# note that viewing the hue channel as grayscale is not very informative
# but the saturation channel clearly shows the difference between colorful
# and netural areas
show_channels(hsv)
```

```python
from skimage.color import hsv2rgb

# increase saturation by muliplying the saturation channel by 2
# (and clipping to avoid values > 1)
s = (hsv[:,:,1] * 2).clip(0,1)
# now replace the saturation channel with the boosted version
hsv[:,:,1] = s
# now convert back to RGB
rgb_enhanced = hsv2rgb(hsv)
show_image(rgb_enhanced)
```
